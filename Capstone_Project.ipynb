### <font color='#4682B4'> 3.  Data Pre-processing </font>
<hr style="height:1.75px;border:none;color:#4682B4;background-color:#4682B4;" />

#### <font color='#4682B4'> 3.1.  Handling missing values </font>

```python
#Checking missing values 

print(data.isnull().sum())
```

```python
#Handling missing values using median 

data['Sleep_Hours']= data['Sleep_Hours'].fillna(data['Sleep_Hours'].median())

data['Stress_Level']= data['Stress_Level'].fillna(data['Stress_Level'].median())
```

```python
#Checking missing values are resolved

print(data.isnull().sum())
```

#### <font color='#4682B4'> 3.2.  Removing duplicates </font>

```python
#checking for duplicate rows

duplicates_count= data.duplicated().sum()
print("Number of duplicate rows:",duplicates_count)
```

```python
#Removing duplicate rows

data=data.drop_duplicates()
```

```python
#Checking duplicate values resolved

duplicates_count= data.duplicated().sum()
print("Number of duplicate rows:",duplicates_count)
```

#### <font color='#4682B4'> 3.3.  Correcting data types </font>

```python
#Gender, Occupation, and Device Type are categorical fields.  
#Converting them from object to category type to improve efficiency and analysis.

categorical_cols = ['Gender', 'Occupation', 'Device_Type']
for col in categorical_cols:
    data[col] = data[col].astype('category')
```

```python
print("\nData types:")
print (data.dtypes)
```

#### <font color='#4682B4'> 3.4.  Creating derived columns  </font>

```python
# A. Total daily screen time
data['Total_Screen_Time'] = data['Daily_Phone_Hours'] + data['Social_Media_Hours']
```

```python
# B. Usage Categorization
def get_usage(hours):
    if hours <= 4: return 'Light'
    elif hours <= 8: return 'Moderate'
    else: return 'Heavy'
data['Usage_Category'] = data['Total_Screen_Time'].apply(get_usage)
```

```python
# C. Stress Categorization
def get_stress(level):
    if level <= 3: return 'Low'
    elif level <= 7: return 'Moderate'
    else: return 'High'
data['Stress_Category'] = data['Stress_Level'].apply(get_stress)
```

```python
# D. Sleep Efficiency (%)
data['Sleep_Efficiency'] = (data['Sleep_Hours'] / 24) * 100
```

```python
# E. Digital Balance Score
data['Digital_Balance_Score'] = (data['Sleep_Hours'] + data['Work_Productivity_Score']) / (data['Total_Screen_Time'] + 1)
```

```python
print(data.columns)
```

#### <font color='#4682B4'> 3.5.  Filtering or aggregating data  </font>

```python
# 1. Filtering (Removing Outliers & Errors) 
# ensuring sleep is realistic (0-18 hours) and screen time is not negative.
data_cleaned = data[(data['Sleep_Hours'] >= 0) & (data['Sleep_Hours'] <= 18)].copy()
data_cleaned = data_cleaned[data_cleaned['Total_Screen_Time'] >= 0]
```

```python
# 2. Final Aggregation (Data Quality Check) 
quality_check = data_cleaned.groupby('Usage_Category').agg({
    'User_ID': 'count',
    'Total_Screen_Time': ['min', 'max', 'mean'],
    'Digital_Balance_Score': 'mean'
}).round(2)
```
